## Search for Citations

```python
import re

# let's assume references is your list of references
# references = ['ref1', 'ref2', 'ref3', ...]

# extract docx content
docx_content = docx2python(draft_doc)

# get the text from the document
text = docx_content.text

# find all citation calls
citation_calls = re.findall(r'\[\d+\]', text)

# for each citation call, find the corresponding reference and create an endnote
for citation_call in citation_calls:
    citation_number = int(citation_call[1:-1])  # strip off the square brackets and convert to int
    print('cite # - ',citation_number)
    corresponding_reference = references[citation_number - 1]  # -1 because list indices are 0-based

    # create an endnote
    endnote = f'----endnote{citation_number}---- {corresponding_reference}\n'
    docx_content.endnotes.append(endnote)
```

```python
matches = []

# for each citation call, check if the author's name appears right before it
for citation_call in citation_calls:
    citation_number = int(citation_call[1:-1])  # strip off the square brackets and convert to int

    corresponding_reference = references[citation_number - 1]  # -1 because list indices are 0-based

    # get the first author's last name from the reference
    first_author_last_name = corresponding_reference.split(',')[0].split()[-1]

    # find the position of the citation call in the text
    citation_pos = text.find(citation_call)

    # get the preceding words in the text
    preceding_words = text[max(0, citation_pos-100):citation_pos].split()

    # check if the author's last name is among the preceding words
    matches.append(first_author_last_name in preceding_words)
```

```python
print('matches - ', matches)
```

## Write back to file

```python
from docx import Document

# create a new document
doc = Document()

# add the text and endnotes to the document
doc.add_paragraph(text)
for endnote in docx_content.endnotes:
    doc.add_paragraph(endnote)

# save the document
doc.save('path/to/new_file.docx')
```
