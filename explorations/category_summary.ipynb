```python
import pandas as pd
```

```python
df = pd.read_excel("step3result copy 2.xlsx") 
df
```

```python
categories = df['category'].unique()
user_question = "Does a diagnosis of a connective tissue disease contribute to a post-dural spinal puncture headache?"
```

```python
from langchain.prompts import (
    ChatPromptTemplate,
    SystemMessagePromptTemplate,
    HumanMessagePromptTemplate,
)
from langchain_openai import AzureChatOpenAI
import os

# remove this later
CHAT = AzureChatOpenAI(
    azure_endpoint="https://nlp-ai-svc.openai.azure.com/",
    openai_api_version="2023-06-01-preview",
    azure_deployment="ChatGPT432k",
    openai_api_type="azure",
    temperature=0,
    model_name="gpt-4-32k",
    openai_api_key=os.environ.get("OPENAI_API_KEY")
)


# CATEGORIZATION_SUMMARY_SYSTEM_TEMPLATE = """I am considering a scoping reivew project for given category I want to understand how the existing literature guides. Write a single detailed summary based on the full texts of the article for each category."""

SUMMARIZE_CATEGORY_TEMPLATE = "I am working  on a scoping review to address this question: '{question}'\n\n Currently, I am summarizing articles by expert-defined categories. All of the articles below were assigned the category {category}. Write a single paragrph summarizing the following journal articles, focusing on my question. Cite each article in the paragraph in APA format."


HUMAN_TEMPLATE = """
Articles sperated by "---":
{articles}
"""

system_message_prompt = SystemMessagePromptTemplate.from_template(SUMMARIZE_CATEGORY_TEMPLATE)
human_message_prompt = HumanMessagePromptTemplate.from_template(HUMAN_TEMPLATE)

categorization_summary_chat_prompt = ChatPromptTemplate.from_messages([system_message_prompt, human_message_prompt])
```

```python
for current_category in categories:
    
    filtered_rows = df[(df['category'] == current_category) & (df['Text'] != "Text not available")]
    # if not filtered_rows.empty:
    #     new_df = pd.concat([new_df, filtered_rows])
    article_texts = []
    for _, row in filtered_rows.iterrows():
        print(row)
        # TODO: double check citation column name
        article_texts.append(f"APA Citation: {row.APA_Citation}\n\n Article text: {row.Text}\n\n --- ")

    text_to_summarize = "\n\n".join(article_texts)
   # text_to_summarize = f"Category of current articles: {current_category} \n\n" + text_to_summarize
    
    result = CHAT.invoke(categorization_summary_chat_prompt.format_prompt(
        question=user_question,
        category=current_category,
        articles=text_to_summarize
        ).to_messages())
    print(result.content)
    print("************")
    
```
