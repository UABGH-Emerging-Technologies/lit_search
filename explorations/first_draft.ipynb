```python
from docx import Document 
```

```python
doc = Document('ExampleStep4Output.docx') 
```

```python
# Initialize an empty string to store the paragraphs
all_paragraphs_text = ""

# Concatenate the text of each paragraph into the string variable
for paragraph in doc.paragraphs:
    all_paragraphs_text += paragraph.text + "\n"

# Now you can use all_paragraphs_text later in your code
print(all_paragraphs_text)
```

```python
user_question = "Does a diagnosis of a connective tissue disease contribute to a post-dural spinal puncture headache?"
```

```python
from langchain.prompts import (
    ChatPromptTemplate,
    SystemMessagePromptTemplate,
    HumanMessagePromptTemplate,
)
from langchain_openai import AzureChatOpenAI
import os

CHAT = AzureChatOpenAI(
    azure_endpoint="https://nlp-ai-svc.openai.azure.com/",
    openai_api_version="2023-06-01-preview",
    azure_deployment="ChatGPT432k",
    openai_api_type="azure",
    temperature=0,
    model_name="gpt-4-32k",
    openai_api_key=os.environ.get("OPENAI_API_KEY")
)


DRAFT_TEMPLATE = """I am working on scoping review to address this question: {question}\n\n Currently, I am drafting a article from the summarizes of the expert-defined categories. Write a final article from the summaries of the categories and make sure to fit into the following format:

Title
A concise and descriptive title that clearly indicates it's a scoping review.

Abstract
A brief summary of the review, including the purpose, methodology, main findings, and conclusions.

Introduction
Background: Contextualize the topic and its importance.
Objectives: Clearly define the objectives of the scoping review.
Rationale: Explain why a scoping review is necessary for this topic.

Methodology
Protocol Development: Describe if a protocol was developed and registered.
Eligibility Criteria: Outline the inclusion and exclusion criteria for studies.
Information Sources: Identify databases and other sources of information.
Search Strategy: Detail the search terms and strategies used.
Selection Process: Explain how studies were selected for inclusion.
Data Charting: Describe the process of extracting and charting the data from the included studies.
Data Items: Identify the specific variables or themes extracted from studies.
Synthesis of Results: Outline the approach for synthesizing and presenting results.

Results
Selection of Sources: Present the results of the search and selection process (often in a PRISMA flow diagram).
Characteristics of Sources: Summarize characteristics of included studies/sources (e.g., type of study, geographical location, population).
Results of Data Charting: Describe the main findings in relation to the review objectives.
Summary of Evidence: Provide a narrative or tabular summary of the evidence, possibly identifying patterns or gaps.

Discussion
Principal Findings: Discuss the main results and how they relate to the review objectives.
Limitations: Acknowledge any limitations of the review.
Implications for Practice, Policy, and Research: Discuss the implications of the findings.
Future Research: Suggest areas for future investigation based on gaps or emerging trends.

Conclusion
Summarize the key findings and their implications in a concise manner.

References
List all the sources cited in the review.

Appendices (if applicable)
Additional materials such as search strategy details, data extraction forms, etc. 
"""


HUMAN_TEMPLATE = """
Content to summarize:
{content}
"""

draft_system_message_prompt = SystemMessagePromptTemplate.from_template(DRAFT_TEMPLATE)

human_message_prompt = HumanMessagePromptTemplate.from_template(HUMAN_TEMPLATE)

draft_chat_prompt = ChatPromptTemplate.from_messages([draft_system_message_prompt, human_message_prompt])
```

```python
result = CHAT.invoke(draft_chat_prompt.format_prompt(
    question=user_question,
    content=all_paragraphs_text
    ).to_messages())
print(result.content)
```
