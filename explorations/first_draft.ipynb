```python
from docx import Document 
```

```python
doc = Document('/data/results/TrialRun/SummarizedArticles100ish.docx') 
```

```python
# Initialize an empty string to store the paragraphs
all_paragraphs_text = ""

# Concatenate the text of each paragraph into the string variable
for paragraph in doc.paragraphs:
    all_paragraphs_text += paragraph.text + "\n\n"

# Now you can use all_paragraphs_text later in your code
print(all_paragraphs_text)
```

```python
user_question = "Does a diagnosis of a connective tissue disease contribute to a post-dural spinal puncture headache?"
```

```python
from langchain.prompts import (
    ChatPromptTemplate,
    SystemMessagePromptTemplate,
    HumanMessagePromptTemplate,
)
from langchain_openai import AzureChatOpenAI
import os

CHAT = AzureChatOpenAI(
    azure_endpoint="https://nlp-ai-svc.openai.azure.com/",
    openai_api_version="2023-06-01-preview",
    azure_deployment="ChatGPT432k",
    openai_api_type="azure",
    temperature=0,
    model_name="gpt-4-32k",
    openai_api_key=os.environ.get("OPENAI_API_KEY")
)


DRAFT_TEMPLATE = """I am working on scoping review to address this question: {question}\n\n 

Currently, I am drafting a article from summaries (with APA-style citations) made from large sets of categorized articles.

Both the categories and summaries were made by experts.

You are an expert medical writer. Your job is to write a draft of the final article with in-text APA-style citations using the summaries of the categories.

Make sure to fit into the following format, citing sources throughout.

# Title
A concise and descriptive title that clearly indicates it's a scoping review.

# Abstract
A brief summary of the review, including the purpose, main findings, and conclusions.

# Introduction

## Background
Contextualize the topic and its importance, including citations from the provided summaries.

## Objectives
Clearly define the objectives of the scoping review.


# Results
Characteristics of Sources: Summarize characteristics of included studies/sources (e.g., type of study, geographical location, population), including citations from the provided summaries.
Results of Data Charting: Describe the main findings in relation to the review objectives, including citations from the provided summaries.
Summary of Evidence: Provide a narrative or tabular summary of the evidence, possibly identifying patterns or gaps, including citations from the provided summaries.

# Discussion
Principal Findings: Discuss the main results and how they relate to the review objectives, including citations from the provided summaries.
Limitations: Acknowledge any limitations of the review.
Implications for Practice, Policy, and Research: Discuss the implications of the findings.
Future Research: Suggest areas for future investigation based on gaps or emerging trends.

# Conclusion
Summarize the key findings and their implications in a concise manner.

# References
Sources are provided in the input content below. Copy all the full citations for sources you cited in the draft review.

# Appendices (if applicable)
Additional materials such as search strategy details, data extraction forms, etc. 

I will add a methodology section later. 

Respond with a draft in markdown format.
"""


HUMAN_TEMPLATE = """
Summaries by categories:
{content}

Please provide a draft of the scoping review article, including in-text citations and a bibliography.
"""

draft_system_message_prompt = SystemMessagePromptTemplate.from_template(DRAFT_TEMPLATE)

human_message_prompt = HumanMessagePromptTemplate.from_template(HUMAN_TEMPLATE)

draft_chat_prompt = ChatPromptTemplate.from_messages([draft_system_message_prompt, human_message_prompt])
```

```python
result = CHAT.invoke(draft_chat_prompt.format_prompt(
    question=user_question,
    content=all_paragraphs_text
    ).to_messages())
print(result.content)
```
