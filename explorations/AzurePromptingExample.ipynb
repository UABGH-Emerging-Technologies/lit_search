```python
from langchain.prompts import PromptTemplate

system_template = """ Help categorize the text file into either of the 3 input categories given by the user. Return 1 category as output."""

system_message_prompt = SystemMessagePromptTemplate.from_template(system_template)

human_template = """
CONTEXT:
{context}

INPUT CATEGORIES:
{category1}, {category2}, {category3}  

OUTPUT:
"""

human_message_prompt = HumanMessagePromptTemplate.from_template(human_template)
```

```python
chat_prompt = ChatPromptTemplate.from_messages([system_message_prompt, human_message_prompt])
result = chat(
chat_prompt.format_prompt(
    category1="Cardiology",category2="Neurology",category3="Anesthesiology", context="extracted_text1.txt"
).to_messages())
```
